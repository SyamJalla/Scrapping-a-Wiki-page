```python
from bs4 import BeautifulSoup
import urllib.request as urllib2
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_towns_in_India_by_population'
```

```python
page =urllib2.urlopen(url)
```

```python
soup = BeautifulSoup(page)
```

```python
soup.prettify
```

```python
all_tables = soup.find_all('table')
```

```python
all_tables
```

```python
right_table = soup.find('table', 'wikitable sortable')
```

```python
right_table
```

```python
A = []
B = []
C = []
D = []
E = []
```

```python
for row in soup.findAll('tr'):
    columns = row.findAll('td')
    if len(columns) == 5:
        A.append(columns[0].find(text=True))
        B.append(columns[1].find(text=True))
        C.append(columns[2].find(text=True))
        D.append(columns[3].find(text=True))
        E.append(columns[4].find(text=True))
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(A)
data['State'] = B
data['Mun_Type'] = C
data['Population'] = D
data['REF'] = E
```
